```python
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
import pandas as pd
```

```python
#load data
data= gutenberg.raw('shakespeare-hamlet.txt') # this is an paragraph of shakesphere maybe poem
```

```python
#data preprocessing
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
data=data.lower()
data# raw text now we will process this text into data
```

```python
#tokenize the text (creating indexes for words)
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])#used to build the vocabulary index
total_words=len(tokenizer.word_index)+1
```

```python
total_words
```

```python
tokenizer.word_index
```

```python
#sentences in our raw text
for sentence in data.split('\n'):
    print(sentence)
```

```python
for line in data.split('\n'): # taking each sentence one by one
    #write the index of words inspite of words
    print(tokenizer.texts_to_sequences([line])[0])
```

```python
#create input sequences
inputseq = []
for line in data.split('\n'): # taking each sentence one by one
    token_list=tokenizer.texts_to_sequences([line])[0]#write the index of words inspite of words
    for i in range(1,len(token_list)):
        n_gram_sequence=token_list[:i+1]
        inputseq.append(n_gram_sequence)
```

```python
inputseq
```

```python
max_len = max([len(x) for x in inputseq])# getting the length of largest sentence to do padding
```

```python
padded_inputseq=pad_sequences(inputseq,maxlen=max_len,padding='pre')
padded_inputseq
```

```python
# so our ouput will be the last word of padded_inputseq 
x= padded_inputseq[:,:-1]
#input will be the input padded_inputseq without last word 
y= padded_inputseq[:,-1]
```

### now our dataset is ready for superived learning and it will be our Multiclass Classification task

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=total_words)# converts class integers into one-hot encoded vectors 
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential([
    Embedding(total_words,100,input_length=max_len),
    LSTM(128),
    Dense(total_words,activation='softmax')
])
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=60)
```

```python
input_text = 'barn. haue'
import numpy as np
#now we will process this test sentence and then predict it by model
for i in range(5):
    text= [input_text]
    token_text=tokenizer.texts_to_sequences([text][0])
    print(token_text)
    padded_token=pad_sequences(token_text,maxlen=14,padding='pre')
    #prediction
    pred=model.predict(padded_token)
    # now we will take the max in them
    pos=np.argmax(pred)
    for word,index in tokenizer.word_index.items():
        if index ==pos:
            input_text=input_text+' '+word
            print(word)
```
